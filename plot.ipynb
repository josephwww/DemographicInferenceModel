# BN Result

```python
import pickle
handle = open('bn_result.pkl', 'rb')
reports = pickle.load(handle)
reports
```

```python
result = {}

for i in range(len(reports)):
    num_movies, attr, prop, accu = reports[i][0], reports[i][1], reports[i][2], reports[i][3]['accuracy']
    if num_movies not in result:
        result[num_movies] = {}
    if attr not in result[num_movies]:
        result[num_movies][attr] = {}
    result[num_movies][attr][prop] = accu
result
```

```python
import numpy as np
import matplotlib.pyplot as plt

for num_movies in result.keys():
    fig = plt.figure()
    plt.title('number of movies: ' + str(num_movies))
    for attr in ('gender', 'age', 'occupation'):
        x, y = list(zip(*result[num_movies][attr].items()))
        plt.errorbar(x, y, label=attr)
    plt.legend(loc='lower right')
        
```

# LR Result

```python
handle = open('lr_result.pkl', 'rb')
reports = pickle.load(handle)
reports
```

```python
result = {}

for i in range(len(reports)):
    user_type, more_or_less, num_movies, attr, prop, accu = reports[i][0], reports[i][1], reports[i][2], reports[i][3], reports[i][4], reports[i][5]['accuracy']
    if user_type not in result:
        result[user_type] = {}
    if more_or_less not in result[user_type]:
        result[user_type][more_or_less] = {}
    if num_movies not in result[user_type][more_or_less]:
        result[user_type][more_or_less][num_movies] = {}
    if attr not in result[user_type][more_or_less][num_movies]:
        result[user_type][more_or_less][num_movies][attr] = {}
    result[user_type][more_or_less][num_movies][attr][prop] = accu
result
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
for user_type in result.keys():
    for more_or_less in result[user_type].keys():
        for num_movies in result[user_type][more_or_less].keys():
            fig = plt.figure()
            plt.title(more_or_less + ' ' + user_type + ' number of movies: ' + str(num_movies))
            for attr in ('gender', 'age', 'occupation'):
                x, y = list(zip(*result[user_type][more_or_less][num_movies][attr].items()))
                plt.errorbar(x, y, label=attr)
            plt.legend(loc='lower right')
        
```
